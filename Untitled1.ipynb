```python
import tensorflow as tf

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import itertools
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import os

print(tf.__version__)
```

```python
def create_custom_dataset(data_dir, batch_size=32, img_size=(224, 224)):
    # Create an ImageDataGenerator for data augmentation and normalization
    datagen = ImageDataGenerator(
        rescale=1./255,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True
    )

    # Create a generator for loading images from the directory
    generator = datagen.flow_from_directory(
        data_dir,
        target_size=img_size,
        batch_size=batch_size,
        class_mode='categorical'  # Use 'binary' for binary classification
    )

    return generator

# Example usage
data_dir = 'week7_data'
batch_size = 32
img_size = (224, 224)
print(os.getcwd())
custom_dataset = create_custom_dataset(data_dir, batch_size, img_size)
```

```python
def create_tf_dataset(generator):
    num_classes = len(generator.class_indices)
    steps_per_epoch = generator.samples // generator.batch_size

    dataset = tf.data.Dataset.from_generator(
        lambda: generator,
        output_signature=(
            tf.TensorSpec(shape=(None, img_size[0], img_size[1], 3), dtype=tf.float32),
            tf.TensorSpec(shape=(None, num_classes), dtype=tf.float32)
        )
    )

    return dataset, steps_per_epoch
tf_dataset, steps_per_epoch = create_tf_dataset(custom_dataset)
```

```python
model= tf.keras.models.Sequential([
    tf.keras.layers.Input((224,224,3)),
    tf.keras.layers.Conv2D(32,3,padding="same",activation="relu"),
    tf.keras.layers.Conv2D(32,3,activation="relu"),
    tf.keras.layers.MaxPooling2D(),
    tf.keras.layers.Dropout(0.20),
    
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(32,activation="relu"),
    tf.keras.layers.Dropout(0.20),
    tf.keras.layers.Dense(2,activation="softmax"),
])
```

```python
model.summary()
```

```python
model.compile(optimizer="adam", loss="categorical_crossentropy",metrics=["acc"])
```

```python
history=model.fit(tf_dataset,epochs=5, steps_per_epoch=steps_per_epoch)
```

```python
fig, ax = plt.subplots(2,1)
ax[0].plot(history.history['loss'], color='b', label="Training Loss")
legend = ax[0].legend(loc='best', shadow=True)

ax[1].plot(history.history['acc'], color='b', label="Training Accuracy")
legend = ax[1].legend(loc='best', shadow=True)
```

```python
cap = cv2.VideoCapture(0)

while True:
    # Read a frame from the webcam
    ret, frame = cap.read()

    # Resize the frame to match the input size of the model
    input_image = cv2.resize(frame, (224, 224))
    input_image = input_image / 255.0  # Normalize pixel values to be between 0 and 1
    input_image = np.expand_dims(input_image, axis=0)  # Add batch dimension

    # Make a prediction using the model
    predictions = model.predict(input_image)

    # Get the predicted class (the index with the highest probability)
    predicted_class = np.argmax(predictions)

    # Display the result on the frame
    cv2.putText(frame, f"Class: {predicted_class}", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
    cv2.imshow("Image Classification", frame)

    # Break the loop if 'q' key is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the webcam and close all windows
cap.release()
cv2.destroyAllWindows()
```
